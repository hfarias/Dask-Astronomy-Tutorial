### Dask.array looks and feels = NumPy

````python
import numpy as np
import dask.array as da
````

````python
np.arange(10)
````

````python
da.arange(10, chunks=5)
````

````python
da.arange(10, chunks=5).compute()
````

## 1D array of ones

Esto divide un array de 15 elementos en tres bloques, cada uno de tamaÃ±o 5:
```
da.ones(15, chunks=(5,))
``` 
    

````python
x = da.ones(15, chunks=(5,))
x.visualize('dask.svg')
````

````python
(x + 1).visualize('dask.svg')
````

````python
(x + 1).sum().visualize('dask.svg')
````

````python
(x + 1).sum().compute()
````

### Algo mas complejo ...

````python
x = da.ones((15, 15), chunks=(5, 5))
x.visualize('dask.svg')
````

````python
(x.dot(x.T)).visualize('dask.svg')
````
